### **Chapter 04. 파이썬 코딩으로 확인하는 최소 제곱**

```python
import numpy as np

x = np.array([2, 4, 6, 8])  # 공부 시간!
y = np.array([81, 93, 91, 97]) #성적(score)

print(x, type(x))   # n-dimensional array (n-차원 배열)
print(y, type(y))
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)

print("x평균:", np.mean(x))
print("y평균:", np.mean(y))
```

```python
# a = 분자 / 분모
분모 = 0
for j in x:
    분모 = 분모 + (j - x_mean)**2

분모=sum([(j - x_mean)**2 for  j in x])
```

```python
분자 = 0
for j in range(0,len(x),1):
    분자 = 분자 + (x[j] - x_mean) * (y[j] - y_mean)
```

```python
a=분자/분모
print("기울기:",a)
```

```python
b=y_mean - (a*x_mean) #y=ax+b  -> b=y-ax
print("y절편:",b)
```

```python
import matplotlib.pyplot as plt


plt.figure() #도화지 한장 준비!
plt.scatter(x,y) #산점도 그리기!

plt.xlabel("Time(h)")
plt.ylabel("Score")

plt.grid() #격자 그리기!
plt.show()
```
