### **Chapter 02. OpenCV 사용하기: HelloCV**

```python
import cv2
import sys

print(cv2.__version__)

img = cv2.imread("lenna.bmp")

if (img is None):
    print("Image load failed!")
    sys.exit()


cv2.namedWindow("image")  # 창의 이름은 자유롭게 작성하면 됨!
cv2.imshow("image", img)
cv2.waitKey()
cv2.destroyAllWindows()
```
