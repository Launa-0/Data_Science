```python
import cv2
import sys


src = cv2.imread("lenna.bmp", cv2.IMREAD_GRAYSCALE)

if (src is None):
    print("Image load failed!")
    sys.exit()

# dst = src + 50         # saturate 연산이 없음!

dst = cv2.add(src, 100)   # saturate 연산이 포함됨!

cv2.namedWindow("src")
cv2.namedWindow("dst")

cv2.imshow("src", src)
cv2.imshow("dst", dst)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2
import sys
import numpy as np

def saturate(value):
    if (value > 255):
        value = 255
    elif (value < 0):
        value = 0

    return value

src = cv2.imread("lenna.bmp", cv2.IMREAD_GRAYSCALE)

if (src is None):
    print("Image load failed!")
    sys.exit()

n = -80

dst = np.empty(shape=src.shape, dtype=src.dtype)

print(dst.dtype)  # uint8 -> unsigend interger 8bit --> 0~255

for y in range(0, src.shape[0], 1):
    for x in range(0, src.shape[1], 1):
        dst[y, x] = saturate(src[y, x] + n)


cv2.namedWindow("src")
cv2.namedWindow("dst")

cv2.imshow("src", src)
cv2.imshow("dst", dst)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
def update(pos):
    dst = cv2.add(src, pos)
    cv2.imshow("dst", dst)

src = cv2.imread("lenna.bmp", cv2.IMREAD_GRAYSCALE)

if (src is None):
    print("Image load failed!")
    sys.exit()

cv2.namedWindow("dst")
cv2.createTrackbar("Brightness", "dst", 0, 100, update)
update(0)

cv2.waitKey()
cv2.destroyAllWindows()
```
