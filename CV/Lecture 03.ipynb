### **Chapter 01. 행렬의 복사**

```python
import cv2
import sys

img = cv2.imread("cat.bmp", cv2.IMREAD_GRAYSCALE)

if (img is None):
    print("Image load failed!")
    sys.exit()


print("Data Type: ", type(img))
print("Shape: ", img.shape)

cv2.namedWindow("cat")
cv2.imshow("cat", img)
cv2.waitKey()
# cv2.destroyWindow("cat")
cv2.destroyAllWindows()
```

```python
img = cv2.imread("cat.bmp", cv2.IMREAD_COLOR)

if (img is None):
    print("Image load failed!")
    sys.exit()


print("Data Type: ", type(img))
print("Shape: ", img.shape)

cv2.namedWindow("cat")
cv2.imshow("cat", img)
cv2.waitKey()
# cv2.destroyWindow("cat")
cv2.destroyAllWindows()
```

```python
import cv2
import sys

img1 = cv2.imread("dog.bmp", cv2.IMREAD_COLOR)

img2 = img1   # 그냥 대입 연사자를 사용해서 복사함!
img3 = img1.copy()  # copy 메서드를 활용해서 복사함!

print(type(img1))
print(img1.shape)

# 통상적으로는 RGB (255, 255, 0)
# BGR @ OpenCV
# yellow <-- Green + Red (0, 255, 255)

# img1[:, :, :] = (255, 255, 0)
img1[:, :, :] = (0, 255, 255)

cv2.namedWindow("image 1")
cv2.namedWindow("image 2")
cv2.namedWindow("image 3")

cv2.imshow("image 1", img1)
cv2.imshow("image 2", img2)
cv2.imshow("image 3", img3)

cv2.waitKey()
cv2.destroyAllWindows()
```

### **Chapter 02. 부분 행렬 추출**

```python
import cv2

img1 = cv2.imread("lenna.bmp", cv2.IMREAD_COLOR)

img2 = img1[200:400, 200:400, :]  # 관심 영역 (Region of Interest, ROI)
img3 = img1[200:400, 200:400, :].copy()

img2 += 50
# img2 = img2 + 50

cv2.namedWindow("Original")
cv2.namedWindow("ROI 1")
cv2.namedWindow("ROI 2")

cv2.imshow("Original", img1)
cv2.imshow("ROI 1", img2)
cv2.imshow("ROI 2", img3)

cv2.waitKey()
cv2.destroyAllWindows()
```
