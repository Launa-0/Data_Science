```python
f=open("Fish.csv") #HDD --(로딩)--> RAM <--(고속도로; 스트림 개방)--CPU
data=csv.reader(f)

h1=next(data) #header 추출
#print(h1)

# (input) x1 :'Weight', x2: 'Length2'
# (output) y: 'Species': Bream ->1, Smelt ->0

bream_weight = [] #비어있는 리스트
bream_length = [] #비어있는 리스트

smelt_weight = [] #비어있는 리스트
smelt_length = [] #비어있는 리스트

for row in data:
  row[1]=float(row[1]) #str -> float
  row[3]=float(row[3]) #str -> float

  if (row[0] == 'Bream'):
    bream_weight.append(row[1])
    bream_length.append(row[3])

  if (row[0] == 'Smelt'):
    smelt_weight.append(row[1])
    smelt_length.append(row[3])

  #print(row)
  #break


f.close()

print("도미의 무게:", bream_weight)
print("도미의 길이:",bream_length)

print("\n빙어의 무게:", smelt_weight)
print("빙어의 길이:",smelt_length)
```

```python
import matplotlib.pyplot as plt

plt.figure() #도화지 한장 준비!

plt.scatter(bream_length,bream_weight, c="blue", s=100, label="Bream") #산점도를 그려주는 함수 호출!
plt.scatter(smelt_length,smelt_weight, c="red", s=100,  label="Smelt")

plt.legend() #n, 전설, 범례

plt.xlabel('Length(cm)')
plt.ylabel('Weight(g)')
plt.grid()

plt.show()
```

### Chapter 02. 첫 번째 머신러닝 프로그램

```python
weight=bream_weight+smelt_weight
length= bream_length+smelt_length

print(weight)
print(length)

print("총 생선 수:",len(weight),"마리")
```

```python
#zip() 함수 + 리스트 내포

x = [[l,w] for l,w in zip(length, weight)]

print(x)
```

```python
print("도미의 총 마리수:",len(bream_weight))
print("빙어의 총 마리수:",len(smelt_weight))
```

```python
#Bream (도미) ----> 1
#Smelt (방어) ----> 0

#y=[1,1,1,1,...,0,0,0]
y=[1]*35+[0]*14

print(y)
```

```python
#scikit-learn 라이브러리
#science  kit -> 과학 + 공구상자
#sklearn

from sklearn.neighbors import KNeighborsClassifier

kn=KNeighborsClassifier()

#1) kn.fit() ---> 학습시킬 때 사용!
#2) kn.score() --> 평가할 때 사용!
#3) kn. predict() --> 주어진 입력 값에 대해서, 정답 값을 예측할 때 사용

kn.fit(x,y) #하이퍼 파라미터(hyper parameter): 개발자가 직접 설정해야 되는 값!
```

```python
result=kn.score(x,y)
print(result) #Accuracy(정확도) =49/49 =1.0
```

```python

plt.figure()

plt.scatter(bream_length,bream_weight, c="blue", s=100, marker='o',label="Bream")
plt.scatter(smelt_length,smelt_weight, c="red", s=100,  marker='s',label="Smelt")


plt.scatter([30,35,13],[600,450,25],c="black",s=100,marker='^',label="New Data")

plt.legend()
plt.xlabel('Length(cm)')
plt.ylabel('Weight(g)')
plt.grid()

plt.show()
```

```python
x_new=[[30,600], #정체불명의 첫번째 물고기
       [35,450], #정체불명의 두번째 물고기
       [13,25]]  #정체불명의 세번째 물고기

y_pred=kn.predict(x_new)
print(y_pred)

for j in range(0,len(y_pred),1):
  if y_pred[j]==1:
    print("도미입니다!")
  else:
    print("빙어입니다!")
```
