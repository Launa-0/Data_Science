###Chapter 05. 텐서플로에서 실행하는 로지스틱 회귀

```python
import numpy as np

x=np.array([2,4,6,8,10,12,14])
y=np.array([0]*3+[1]*4)

print("공부시간:",x)
print("합격여부:",y)

print(type(x),type(y))
```

```python
from tensorflow import keras
from keras import Sequential, Input
from keras.layers import Dense

model=Sequential()
model.add(Input(shape=(1,))) #입력층(Input Layer), 특징의 갯수를 1로 설정!
model.add(Dense(units=1,activation="sigmoid")) #출력값의 갯수를 1로 설정, y=sigmoid(w*x+b)
```

```python
#손실 함수 (loss function) ->loss
#Optimizzeer -> 경사 하강법 알고리즘 지정!
model.compile(loss="binary_crossentropy",optimizer="sgd")
#sgd -> stochastic gradient descent

#에포크(epoch) ->  문제집을 몇번 반복해서 풀게(학습하게) 할 것인지 결정
model.fit(x,y,epochs=500)
```

```python
import matplotlib.pyplot as plt

x_new=np.arrange(0,14,0.5)

plt.figure() #도화지 1장 준비!

plt.scatter(x,y,c="red",s=150)

plt.plot(x_new,model.predict(x_new),marker="s",c="black")

plt.xlabel("Study Time(x)")
plt.ylabel("Pass or Failure(y)")
plt.grid()

plt.show()
```
